```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_AllStar = pd.read_csv("PlusMinusForAllStars.csv")
df_NonStar = pd.read_csv("PlusMinusNonAllstar.csv")
```

```python
df_AllStar
```

```python
df_NonStar
```

```python
df_preStar = df_AllStar["Pre-Covid"]
df_postStar = df_AllStar["Post-Covid"]
nameStar = df_AllStar["Name"]
```

```python
pre_covid = plt.plot(nameStar,df_preStar, alpha = 1, linestyle='-', marker='o')
post_covid = plt.plot(nameStar,df_postStar, alpha = 1, linestyle='-', marker='o')
plt.xticks(rotation=90)
plt.legend([pre_covid[0], post_covid[0]], ['Pre Covid', 'In Bubble'])
plt.figure(figsize=(500,100))
plt.plot(figsize = (100, 100))
```

```python
df_betterStar = df_AllStar[df_AllStar["Post-Covid"] > df_AllStar["Pre-Covid"]]
df_betterStar
```

```python
df_worseStar = df_AllStar[df_AllStar["Post-Covid"] < df_AllStar["Pre-Covid"]]
df_worseStar
```

```python
pre_covid = plt.plot(df_betterStar["Name"],df_betterStar["Pre-Covid"], alpha = 1, linestyle='-', marker='o')
post_covid = plt.plot(df_betterStar["Name"],df_betterStar["Post-Covid"], alpha = 1, linestyle='-', marker='o')
plt.xticks(rotation=90)
plt.title("Observed Improvement in +/- for All-Stars")
plt.xlabel("Name")
plt.ylabel("Observed +/-")
plt.legend([pre_covid[0], post_covid[0]], ['Pre-Covid', 'In Bubble'])
plt.figure(figsize=(500,100))
plt.show()
```

```python
pre_covid = plt.plot(df_worseStar["Name"],df_worseStar["Pre-Covid"], alpha = 1, linestyle='-', marker='o')
post_covid = plt.plot(df_worseStar["Name"],df_worseStar["Post-Covid"], alpha = 1, linestyle='-', marker='o')
plt.title("Observed Regression in +/- for All-Stars")
plt.xlabel("Name")
plt.ylabel("Observed +/-")
plt.xticks(rotation=90)
plt.legend([pre_covid[0], post_covid[0]], ['Pre-Covid', 'In Bubble'])
plt.figure(figsize=(500,100))
plt.show()
```

```python
df_new = df_AllStar.drop(columns = ["Name"])
boxplot = df_new.boxplot()
plt.title("All-Star")
plt.ylabel("Observed +/-")
plt.show()
```
